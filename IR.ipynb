```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

text = "Tokenization, stop word removal, and stemming are essential steps in preprocessing unstructured text data."
print("Original Text:")
print(text)
text = text.lower()
text = re.sub(r'[^\w\s]', '', text)

tokens = text.split()
print("\nAfter Tokenization:")
print(tokens)

stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]
print("\nAfter Stop-word Removal:")
print(filtered_tokens)

stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]
print("\nAfter Stemming:")
print(stemmed_tokens)

print("\nVocabulary Size Before Preprocessing:", len(set(tokens)))
print("Vocabulary Size After Preprocessing:", len(set(stemmed_tokens)))
```

```python
import numpy as np
from collections import Counter
import math

doc1 = "machine learning is fun"
doc2 = "learning machine is important"

words1 = doc1.lower().split()
words2 = doc2.lower().split()

vocab = list(set(words1 + words2))

counter1 = Counter(words1)
counter2 = Counter(words2)

vec1 = [counter1[word] for word in vocab]
vec2 = [counter2[word] for word in vocab]

def cosine_similarity(v1, v2):
    dot = np.dot(v1, v2)
    mag1 = math.sqrt(np.dot(v1, v1))
    mag2 = math.sqrt(np.dot(v2, v2))
    return dot / (mag1 * mag2)

print("Vocabulary:", vocab)
print("Vector 1:", vec1)
print("Vector 2:", vec2)
print("Cosine Similarity:", cosine_similarity(vec1, vec2))
```

```python
import math

D1 = "information retrieval system vector"
D2 = "information retrieval similarity measure"
D3 = "vector space model retrieval"
Q  = "information retrieval vector"

docs = [D1, D2, D3]
names = ["D1", "D2", "D3"]

words = list(set((D1 + " " + D2 + " " + D3 + " " + Q).split()))

print("Binary Term Incidence Matrix:\n")

print("Word List:", words)
print()

for i in range(3):
    print(names[i], end=" : ")
    for w in words:
        if w in docs[i].split():
            print(1, end=" ")
        else:
            print(0, end=" ")
    print()

def cosine(d, q):
    v1 = []
    v2 = []

    for w in words:
        v1.append(d.split().count(w))
        v2.append(q.split().count(w))

    dot = 0
    for i in range(len(v1)):
        dot += v1[i] * v2[i]

    norm1 = math.sqrt(sum(x*x for x in v1))
    norm2 = math.sqrt(sum(x*x for x in v2))

    if norm1 == 0 or norm2 == 0:
        return 0

    return dot / (norm1 * norm2)

def jaccard(d, q):
    set1 = set(d.split())
    set2 = set(q.split())

    intersection = len(set1 & set2)
    union = len(set1 | set2)

    return intersection / union

print("\nSimilarity Values:\n")

for i in range(3):
    cos = round(cosine(docs[i], Q), 4)
    jac = round(jaccard(docs[i], Q), 4)
    print(names[i], " -> Cosine:", cos, " Jaccard:", jac)
```
